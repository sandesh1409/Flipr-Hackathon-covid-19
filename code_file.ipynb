```python
import pandas as pd
df = pd.read_excel('/home/sandynote/Desktop/Flipr Hackathon/Train_dataset.xlsx')
df.head()
```

```python
df.columns
```

```python
df1 = df.drop(['people_ID','Heart rate','Designation','Name','cardiological pressure'],axis=1)
```

```python
numeric_df = df1.select_dtypes(include=['int64', 'float64'])
categorical_df = df1.select_dtypes(include='object')
```

```python
numeric_df.shape, categorical_df.shape
```

```python
#Fill Null values

numeric_df = numeric_df.fillna(0)
categorical_df = categorical_df.fillna('Unknown')
```

```python
df= pd.concat([categorical_df,numeric_df], axis=1)
df.head()
```

```python
df['Pulmonary score'] = df['Pulmonary score'].map(lambda x:x.lstrip('<')).astype(int)
```

```python
df.info()
```

```python
df.head()
```

```python
y_train = df['Infect_Prob'];
x_train = df.drop(['Infect_Prob'],axis=1);
                  
```

```python
#bulding a test data
dff = pd.read_excel('/home/sandynote/Desktop/Flipr Hackathon/Test_dataset.xlsx')
dff.head()
```

```python
dff1 = dff.drop(['people_ID','Heart rate','Designation','Name','cardiological pressure'],axis=1)
```

```python

numeric_dff = dff1.select_dtypes(include=['int64', 'float64'])
categorical_dff = dff1.select_dtypes(include='object')

#Fill Null values
numeric_dff = numeric_dff.fillna(0)
categorical_dff = categorical_dff.fillna('Unknown')

dff2= pd.concat([categorical_dff,numeric_dff], axis=1)

dff2['Pulmonary score'] = dff2['Pulmonary score'].map(lambda x:x.lstrip('<')).astype(int)
x_test = dff2
```

```python
import numpy as np
categorical_features_indices = np.where(x_train.dtypes == 'object')[0]
```

```python
#importing library and building model
from catboost import CatBoostRegressor
model=CatBoostRegressor(iterations=100, depth=3,  loss_function='RMSE')
model.fit(x_train, y_train,cat_features=categorical_features_indices)
```

```python
y_pred = model.predict(x_test)
y_pred = pd.DataFrame(y_pred)
y_pred.columns = ['predicted_probability']
people_id = dff.people_ID
```

```python
answer = pd.concat([people_id,y_pred],axis=1)
```

```python
from IPython.display import HTML
import base64  


def create_download_link( df, title = "Download CSV file", filename = "data.csv"):  
    csv = df.to_csv(index =False)
    b64 = base64.b64encode(csv.encode())
    payload = b64.decode()
    html = '<a download="{filename}" href="data:text/csv;base64,{payload}" target="_blank">{title}</a>'
    html = html.format(payload=payload,title=title,filename=filename)
    return HTML(html)

create_download_link(answer)
```
